```python
import pandas as pd

df = pd.read_csv("Walmart.csv")
```

```python
df.head()
```

```python

df['transaction_date'] = pd.to_datetime(
    df['transaction_date'],
    errors='coerce',
    format='mixed'
)
```

```python
df['transaction_date'].isnull().sum()
```

```python
df.info()
```

```python
df['transaction_date'] = df['transaction_date'].dt.strftime('%Y-%m-%d %H:%M:%S')
```

```python
df.to_csv("Walmart_clean.csv", index=False)
```
